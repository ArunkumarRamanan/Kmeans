# K-Means Clustering

```python
import pandas as pd
import numpy as np #For Linear Algebra
import matplotlib.pyplot as plt
%matplotlib inline
from ggplot import * 
from sklearn.cluster import KMeans 
from collections import Counter
from matplotlib import style
style.use("ggplot")
```

```python
df=pd.read_csv("data_1024.csv",sep='\t')
```

we're storing feature data from our dataframe `df`, in the `f1` and `f2` arrays. We combine this into a feature matrix `X` before entering it into the algorithm.

```python
f1 = df['Distance_Feature'].values
f2 = df['Speeding_Feature'].values
X=np.matrix(list(zip(f1,f2)))
```

```python
print (ggplot(df, aes('f1', 'f2')) + geom_point(colour='steelblue'))
```

```python
#Starting by choosing K=2  
kmeans = KMeans(n_clusters=2).fit(X)
y_kmeans = kmeans.predict(X)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
c = Counter(labels)
```

```python
print ("centroids : ")
print (centroids)
print ("labels : ")
print (labels)
```

```python
plt.scatter([X[:, 0]], [X[:, 1]], c=y_kmeans, s=50, cmap='viridis')

plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=200, alpha=0.5);
```

```python
#Starting by choosing K=4
kmeans = KMeans(n_clusters=4).fit(X)
y_kmeans = kmeans.predict(X)
```
